```python
import requests
import pandas as pd
import json
```

```python
url_nominatim = "https://nominatim.openstreetmap.org/search"
```

```python
villes = [
    "Mont Saint Michel",
    "St Malo",
    "Bayeux",
    "Le Havre",
    "Rouen",
    "Paris",
    "Amiens",
    "Lille",
    "Strasbourg",
    "Chateau du Haut Koenigsbourg",
    "Colmar",
    "Eguisheim",
    "Besancon",
    "Dijon",
    "Annecy",
    "Grenoble",
    "Lyon",
    "Gorges du Verdon",
    "Bormes les Mimosas",
    "Cassis",
    "Marseille",
    "Aix en Provence",
    "Avignon",
    "Uzes",
    "Nimes",
    "Aigues Mortes",
    "Saintes Maries de la mer",
    "Collioure",
    "Carcassonne",
    "Ariege",
    "Toulouse",
    "Montauban",
    "Biarritz",
    "Bayonne",
    "La Rochelle"
    ]
```

```python
# Fonction pour obtenir les données GPS d'une ville
def get_gps_data(city):
  
  # Paramètres de la requête
  params = {
    "q": city,
    "format": "json",
  }
  
  # Envoi de la requête et récupération de la réponse
  response = requests.get(url_nominatim, params=params)

  # Vérification du statut de la requête
  if response.status_code == 200:
    # Analyse de la réponse JSON
    data = response.json()

    # Extraction des données GPS
    latitude = data[0]["lat"]
    longitude = data[0]["lon"]

    # Affichage des données GPS
    print(f"Ville : {city}")
    print(f"Latitude : {latitude}")
    print(f"Longitude : {longitude}")
  else:
    # Affichage d'un message d'erreur
    print(f"Erreur : {response.status_code}")

for ville in villes:
  get_gps_data(ville)
```
