```python
import json
import csv


# Fonction pour obtenir les données GPS d'une ville
def get_gps_data(city):
    # Paramètres de la requête
    params = {
        "q": city,
        "format": "json",
    }

    # Envoi de la requête et récupération de la réponse
    response = requests.get(url_nominatim, params=params)

    # Vérification du statut de la requête
    if response.status_code == 200:
        # Analyse de la réponse JSON
        data = response.json()

        # Extraction des données GPS
        if data: # Vérifier si data n'est pas None
            latitude = data[0]["lat"]
            longitude = data[0]["lon"]
        else:
           latitude = ""  # Remplacer None par une chaîne vide
           longitude = ""  # Remplacer None par une chaîne vide

        # Affichage des données GPS
        print(f"Ville : {city}")
        print(f"Latitude : {latitude}")
        print(f"Longitude : {longitude}")
    else:
        # Affichage d'un message d'erreur
        print(f"Erreur : {response.status_code}")

# Itérer sur les villes
for ville in villes:
    # Extraire les données GPS
    data = get_gps_data(ville)
    
    # Préparer les données pour le fichier CSV
    if data:
        csv_data = [ville, data[0]["lat"], data[0]["lon"]]
    else:
        csv_data = [ville, None, None]

    # Ouvrir le fichier CSV en mode append
    with open("gps_data.csv", "a", newline="") as csvfile:
        # Créer un objet writer
        writer = csv.writer(csvfile)

        # Écrire la ligne de données
        writer.writerow(csv_data)

  

   
```

```python
import json
import csv
import requests # Bibliothque supplémentaire pour les requêtes API

# Fonction pour obtenir les données GPS d'une ville (utilisant une API de géocodage fictive)
def get_gps_data(lieu):
    # Paramètres de la requête
    params = {
        "q": city,
        "format": "json",
    }

    # Envoi de la requête et récupération de la réponse
    response = requests.get(url_nominatim, params=params)

    # Analyser la réponse JSON et extraire les coordonnées GPS
    if response.status_code == 200:
        data = response.json()
        if data["results"]:
            latitude = data["results"][0]["geometry"]["location"]["lat"]
            longitude = data["results"][0]["geometry"]["location"]["lng"]
            return latitude, longitude
        else:
            return None, None
    else:
        print(f"Erreur lors de la requête API pour {lieu}: {response.status_code}")
        return None, None

# Itérer sur les lieux
for lieu in lieux:
    # Extraire les données GPS
    latitude, longitude = get_gps_data(lieu)

    # Préparer les données pour le fichier CSV
    csv_data = [lieu, latitude, longitude]

    # Ouvrir le fichier CSV en mode append
    with open("gps_data.csv", "a", newline="") as csvfile:
        # Créer un objet writer
        writer = csv.writer(csvfile)

        # Écrire la ligne de données
        writer.writerow(csv_data)
```
